Security Allocation
In this project we will finds an optimal allocation of stocks in a portfolio satisfying a minimum expected return. The problem is solved by using qualdradic programming method. We will use the covxopt library to solve for optimal weight of stocks. Due to hard to compile the cvxopt library in to Anaconda, we use Python(x,y) for this project. We will webscrap the stock data from yahoo finance.

```python
# getting stock data from Yahoo

from pandas import Series, DataFrame
import pandas as pd
import numpy as np
import pandas.io.data as web
import time

def StockReturn(portfolio,start = '1/1/2010',end = time.strftime("%m/%d/%Y")):
    all_data={}
    for ticker in portfolio:
        all_data[ticker] = web.get_data_yahoo(ticker, start, end)
    price = DataFrame({tic : data['Adj Close'] for tic, data in all_data.iteritems()})
    price = price.dropna()
    returns = price.pct_change()
    R_DF = returns.dropna()
    return R_DF, price
```

```python
class statistics:
    def __init__(self,portfolio):
        self.portfolio = portfolio
        self.returns = StockReturn(self.portfolio)[0]
        self.n = size(portfolio)
        
    def MeanR(self):
        Rp = self.returns.as_matrix()
        return Rp.mean(0).reshape(self.n,1)
    
    def covR(self):
        covR_DF = self.returns.cov()
        return covR_DF.as_matrix()
    
    def varR(self):
        covR = covR(self)
        return covR.diagonal().reshape(self.n,1)
```

solves the QP, where w is the allocation of the portfolios:
minimize   u2w'Pw + q'w
subject to Gw <= h
           Aw == b

```python
%matplotlib inline
import matplotlib.pyplot as plt
from cvxopt import matrix, solvers
solvers.options['show_progress'] = False

def optimal(portfolio, m = 500, Rf = 0.03):
    pf = statistics(portfolio)
    n = pf.n
    #Rtn = pf.returns.as_matrix()
    covR = pf.covR()
    MeanR = pf.MeanR()
    
    mup = np.arange(min(MeanR),max(MeanR)+(max(MeanR)-min(MeanR))/m,(max(MeanR)-min(MeanR))/m)
    weights = np.zeros((mup.shape[0],n))
    risk = np.zeros((mup.shape[0],1))
    P_return = np.zeros((mup.shape[0],1))
    
    P = matrix(covR)
    # w = n by 1 vector
    q = matrix(np.zeros((n, 1)), tc='d')
    # inequality constraints Gw <= h
    # captures the constraints (MeanR'w = mup) and (1w = 0)    
    # Create constraint matrices
    G = matrix(0.0, (n,n))
    G[::n+1] = -1.0   # negative n x n identity matrix
    h = matrix(0.0, (n ,1))
    A = matrix(np.concatenate((MeanR.T, np.ones((1,n))),0))
        
    ### solve
    for i in range(weights.shape[0]):
        b = matrix(array((mup[i], 1.0)).reshape(2,1))
        tmp1 = solvers.qp(P, q, G, h, A, b)['x'].T
        tmp = 1- sum(tmp1[0:n-1])
        weights[i,] = tmp1
        weights[i,n-1] = tmp
        risk[i] = (np.dot(np.dot(weights[i,],covR),weights[i,].T))**(0.5)
        P_return[i] = np.dot(weights[i,].T, MeanR)
    
    ### optimized weight based on sharp ratio
    sds_v = risk.reshape(risk.shape[0])
    sharp = (mup - Rf/253)/sds_v
    indx = (sharp == max(sharp))
    opt_w = weights[indx,]
    
    return weights, risk, P_return, opt_w, indx
   
```

```python
def rand_weights(n):
    ''' Produces n random weights that sum to 1 '''
    k = np.random.rand(n)
    return k / sum(k)
```

```python
def random_portfolio(returns):
    ''' 
    Returns the mean and standard deviation of returns for a random portfolio
    '''

    p = np.asmatrix(np.mean(returns, axis=1))
    w = np.asmatrix(rand_weights(returns.shape[0]))
    C = np.asmatrix(np.cov(returns))
    
    mu = w * p.T
    sigma = np.sqrt(w * C * w.T)
    
    # This recursion reduces outliers to keep plots pretty
    if sigma > 2:
        return random_portfolio(returns)
    return mu, sigma
```

```python
returns = returns.T
n_portfolios = 1000
means, stds = np.column_stack([
    random_portfolio(returns) 
    for _ in xrange(n_portfolios)
])
```
